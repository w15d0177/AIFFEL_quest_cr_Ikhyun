```python
# 사각형 넓이를 구하는 클래스 완성!
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
         base, height = map(int, input('밑변, 높이를 입력하세요. 예시: 밑변,높이\n >>> ').split(','))
         area = base * height
         result = '평행사변형의 넓이는 : ' + str(area)
         return result

    def trape(self):
        base1, base2, height = map(int, input('윗변, 아랫변, 높이를 입력하세요. 예시 : 윗변,아랫변,높이\n >>> ').split(','))
        area = (base1 + base2) * height / 2
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result

a = Square()  # 객체 생성 & 2, 3을 각각 입력해 봅시다.
```

```python
a.par()
```

```python
a.trape()
```

```python
# 문제 3-3. 클래스 업데이트  
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요! 

class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트
        self.price_sum = 0
        self.temp = ""  # 음료 온도 초기화

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(f"{i + 1}. {self.menu[i]} : {self.price[i]} 원")

    # 주문 메서드
    def menu_select(self):
        print()  # 줄 바꿈
        n = int(input("음료를 번호를 입력하세요 : "))  # 음료 번호 입력

        # 선택 음료의 가격을 합계에 추가
        self.price_sum = self.price[n-1]

        # 음료 온도 물어보기
        t = 0  # 기본값을 넣어주고
        while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

            if t == 1:
                self.temp = "HOT"  # 온도 속성 생성
            elif t == 2:
                self.temp = "ICE"  # 온도 속성 생성
            else:
                print("다시 입력하세요.\n")

        # 주문 결과 출력
        print(f"추가 주문 음료: {self.temp} {self.menu[n-1]}, {self.price[n-1]}원")

        # 주문 추가
        self.order_menu.append(f"{self.temp} {self.menu[n-1]}")
        self.order_price.append(self.price[n-1])
        self.price_sum += self.price[n-1]

        print(f"{self.temp} {self.menu[n-1]} : {self.price[n-1]} 원")
        print(f"합계 : {self.price_sum} 원")

    # 추가 주문 또는 지불
    def additional_order(self):
        while True:
            print()  # 줄 바꾸기
            n = int(input("추가 주문 음료 번호를 입력하거나, 지불을 원하시면 0을 누르세요 : "))  # 추가 주문 또는 지불

            if n == 0:  # 지불을 입력하면
                print("주문이 완료되었습니다.")
                print(self.order_menu, self.order_price)  # 확인을 위한 리스트를 출력
                break
            elif 1 <= n <= len(self.menu):
                # 음료 온도 물어보기
                t = 0  # 기본값을 넣어주고
                while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

                    if t == 1:
                        self.temp = "HOT"  # 온도 속성 생성
                    elif t == 2:
                        self.temp = "ICE"  # 온도 속성 생성
                    else:
                        print("다시 입력하세요.\n")

                # 주문 리스트에 추가
                self.order_price.append(self.price[n-1])  # 가격 리스트에 추가
                self.order_menu.append(f"{self.temp} {self.menu[n-1]}")  # 주문 리스트에 추가
                self.price_sum += self.price[n-1]  # 합계 금액에 더하기

                # 주문 결과 출력
                print(f"추가 주문 음료: {self.temp} {self.menu[n-1]}, {self.price[n-1]}원\n합계: {self.price_sum} 원")
            else:  # 없는 번호를 입력할 때
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    # 지불 메서드
    def pay(self):
        print(f"총 합계 금액은 {self.price_sum}원 입니다.")

        self.pay_method = 0  # 기본값 넣어주고
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input("현금 결제는 1번, 카드 결제는 2번을 입력해 주세요: "))

            if self.pay_method == 1:
                print("직원을 호출하겠습니다.")
            elif self.pay_method == 2:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
            else:
                print("다시 결제를 시도해 주세요.")

    # 주문서 출력 메서드
    def table(self):
        outline1 = '⟝' + '-' * 30 + '⟞'
        outline2 = '|' + ' ' * 31 + '|'

        print(outline1)
        for i in range(5):
            print(outline2)

        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ':', self.order_price[i])

        print('합계금액:', self.price_sum)

        for i in range(5):
            print(outline2)
        print(outline1)
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
